**MINOR PROJECT BY MULTI-LINEAR REGRESSION**

```python
# IMPORTING ALL REQ LIBRARIES

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

**DATA CLEANING**

```python
df_minor = pd.read_csv('https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv')
df_minor
```

```python
# selecting ONLY INT & FLOAT VALUES :

df_minor = df_minor.select_dtypes(include = ['int64' , 'float64'])
df_minor
```

**NORMAL DATA VISUALISATION**

```python
# DATA VISUALISATION WITHOUT MULTI-LINEAR REGRESSION

sns.distplot(df_minor['Profit'])
plt.show()
```

```python
# SHAPE OF DATAFRAME

df_minor.shape
```

```python
# INPUT to MULTI-LINEAR REGRESSION (without PROFITS)
x = df_minor.iloc[:,0:3].values

# OUTPUT to MULTI LINEAR REGRESSION (only PROFITS)
y = df_minor.iloc[:,-1].values
```

```python
# IMPORTING MODEL
# by default the split is 75:25 

from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# TRAINING THE MACHINE 

model1 = LinearRegression()
model1.fit(x_train,y_train)
```

```python
# PREDICTING OUTPUT

y_pred = model1.predict(x_test)
y_pred 
```

```python
# ACTUAL PROFIT v/s PREDICTED PROFIT

df_m = pd.DataFrame({'ACTUAL PROFIT' : y_test , 'PREDICTED PROFIT' : y_pred})
df_m
```

**DATA VISUALISATION**

```python
# ACTUAL v/s PREDICTED -> NORMAL PLOT

df_m.plot(figsize=(20,8))
plt.show()
```

```python
# ACTUAL v/s PREDICTED ->  BAR GRAPH

df_m.plot(figsize=(20,8), kind = 'bar')
plt.show()
```

```python
sns.regplot(x = 'ACTUAL PROFIT' , y = 'PREDICTED PROFIT' , data = df_m)
plt.show()
```

**REGRESSION SCORE**

```python
# REGRESSION SCORE
from sklearn.metrics import r2_score 
r2_score(y_test,y_pred)
```

**MODEL PERSISTANCE**

```python
# MODEL PERSISTANCE (creating a file to share and use it elsewhere as a model)
import pickle
```

```python
# SERIALISATION STEP
pickle.dump(model1,open("minor project.pkl","wb")) 

# DESERIALISATION STEP
reload_model = pickle.load(open("minor project.pkl","rb"))
```
